```python
# Two lists l1, l2 of same length : Sum of respective elements from both list
l1 = list(range(10, 20, 2))
l2 = list(range(31, 41, 2))
```

```python
print(len(l1), len(l2))
print(l1)
print(l2)
```

```python
# 1. nested for loop
l_sum = []
for i in range(len(l1)):
  for j in range(len(l2)):
    if i == j:
      s = l1[i] + l2[j]
      l_sum.append(s)
```

```python
l_sum
```

```python
# 2. zip()
l_sum = []
for i, j in zip(l1, l2):
  l_sum.append(i+j)
print(l_sum)
```

```python
for i in zip(l1, l2):
  print(i)
```

```python
# 3. range() function
l_sum = []
for i in range(len(l1)):
  l_sum.append(l1[i]+l2[i])
print(l_sum)
```

```python
# 4. enumerate() function
l_sum = []
for idx, j in enumerate(l1):
  l_sum.append(l1[idx]+l2[idx])
print(l_sum)
```

```python
# 5. lambda function
list(map(lambda x, y : x+y, l1, l2))
```
